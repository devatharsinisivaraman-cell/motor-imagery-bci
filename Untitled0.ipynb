```python
# ----------------------------
# Step 1: Install Libraries
# ----------------------------
!pip install mne scikit-learn matplotlib seaborn

# ----------------------------
# Step 2: Import Libraries
# ----------------------------
import mne
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report, confusion_matrix, precision_score, recall_score

# ----------------------------
# Step 3: Download EEG Data
# ----------------------------
# Using Subject 1, Run 4 (Motor Imagery: left vs right hand)
!wget -q https://physionet.org/files/eegmmidb/1.0.0/S001/S001R04.edf

# ----------------------------
# Step 4: Load EEG Data
# ----------------------------
raw = mne.io.read_raw_edf("S001R04.edf", preload=True)
raw.pick_types(eeg=True)          # Keep EEG channels only
raw.filter(1., 40.)               # Band-pass filter 1-40 Hz

print("EEG Channels:", raw.ch_names[:10])  # Show first 10 channels
print(raw)

# ----------------------------
# Step 5: Extract Events from Annotations
# ----------------------------
events, event_id = mne.events_from_annotations(raw)
print("Event IDs:", event_id)      # T0=rest, T1=left, T2=right

# Keep only T1 and T2 for classification
epochs = mne.Epochs(raw, events, event_id={'T1':2,'T2':3},
                    tmin=0, tmax=4, baseline=None, preload=True)
labels = epochs.events[:, -1]     # last column is label
print("Number of epochs:", len(epochs))
print("Labels:", np.unique(labels))

# ----------------------------
# Step 6: Feature Extraction (mean + std per channel)
# ----------------------------
X = epochs.get_data()              # shape: (n_epochs, n_channels, n_times)
features = np.hstack((X.mean(axis=2), X.std(axis=2)))
print("Feature shape:", features.shape)

# ----------------------------
# Step 7: Train/Test Split and Scaling
# ----------------------------
X_train, X_test, y_train, y_test = train_test_split(
    features, labels, test_size=0.2, random_state=42)

scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# ----------------------------
# Step 8: Train Classifier (SVM)
# ----------------------------
clf = SVC(kernel='rbf', C=1, gamma='scale')
clf.fit(X_train, y_train)
y_pred = clf.predict(X_test)

# ----------------------------
# Step 9: Evaluate Model
# ----------------------------
print("Classification Report:\n", classification_report(y_test, y_pred))
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=['Left Hand (T1)','Right Hand (T2)'],
            yticklabels=['Left Hand (T1)','Right Hand (T2)'])
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()

# ----------------------------
# Step 10: Visualize EEG Signals (Example Epochs)
# ----------------------------
class_labels = {2: "Left Hand (T1)", 3: "Right Hand (T2)"}

for label in np.unique(labels):
    epoch_idx = np.where(labels == label)[0][0]  # pick first sample
    sample = X[epoch_idx]  # shape: (n_channels, n_times)

    plt.figure(figsize=(12, 6))
    for i in range(min(10, sample.shape[0])):  # plot first 10 channels
        plt.plot(sample[i] + i*50, label=f"Ch {i+1}")  # shift for clarity
    plt.title(f"EEG signals for {class_labels[label]}")
    plt.xlabel("Time (samples)")
    plt.ylabel("Amplitude (shifted)")
    plt.show()

# ----------------------------
# Step 11: Per-Class Accuracy (Bar Plot)
# ----------------------------
precisions = []
recalls = []
for c in np.unique(labels):
    precisions.append(precision_score(y_test, y_pred, labels=[c], average='macro'))
    recalls.append(recall_score(y_test, y_pred, labels=[c], average='macro'))

plt.figure(figsize=(6,4))
x = np.arange(len(np.unique(labels)))
plt.bar(x - 0.2, precisions, width=0.4, label='Precision')
plt.bar(x + 0.2, recalls, width=0.4, label='Recall')
plt.xticks(x, ['Left Hand (T1)', 'Right Hand (T2)'])
plt.ylabel("Score")
plt.title("Model Performance per Class")
plt.legend()
plt.show()
```
